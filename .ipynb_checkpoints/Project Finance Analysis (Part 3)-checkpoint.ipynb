In this notebook, you will do the following:

```python
# Step 1: Import your libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Read your CSV from Part II and make sure your index is parsed as a date
df = pd.read_csv('finance_data_extended.csv', parse_dates=True, index_col=0)
df
```

### Step 3: Slice your DataFrame to cover only 2009-2019 
The prices in 2020 will not be reliable, since we've the unexpected event that is COVID-19 which has done very bad things to the economy.

As such, let's just look at the past ten years of data, i.e. from 2009-01-01 to 2019-01-01. Since your index is already a DateTime object, you can <strong>slice</strong> them using strings.

```python
# Step 3: slice your DataFrame by indexing the dates
new_df = df.loc['2009']
for i in range(9):
    new_df = pd.concat([new_df,df.loc[str(2010 + i)]])
new_df
    
```

### Step 4: Sort your DataFrame by Index
If you noticed, your DataFrame currently starts with the latest date, followed by the oldest date. 

This won't do because we will need our data to be chronological order, i.e. oldest to latest, for our subsequent calculations.

```python
# Step 4: sort your DataFrame's index
new_df.sort_index(inplace=True)
new_df
```

### Step 5: Calculate the interday percentage change in close price
Now for our first analysis - calculating the interday percentage change in close price. We do this to see how the prices change over time in terms of percentages. 

```python
# Step 5: Create a new column that contains the interday percentage change
new_df['interday_close_percent'] = new_df['4. close'].pct_change()
new_df
```

### Step 6: Implement a buy-and-hold strategy
An underrated strategy, and sometimes said to outperform many other strategies, the buy-and-hold strategy is exactly what it sounds like - you buy-and-hold onto the stocks without doing anything.

```python
initial = new_df['4. close']['2009-01-02'] * 100

portfolio = [initial]

for i in new_df['interday_close_percent'][1:]:
    new_value = portfolio[-1] * (1+i)
    portfolio.append(new_value)

new_df['portfolio_hold_loop'] = portfolio
```

```python
new_df
```

```python
# Step 7: Create a column containing a 50-day window SMA
new_df['50-day Moving Average'] = new_df['4. close'].rolling(50).mean()
new_df
```

```python
# Step 8: Plot your close price and 50-day SMA
plt.figure(figsize = (16,8))
plt.plot(new_df['4. close'], c = 'orange', label = 'Close Price')
plt.plot(new_df['50-day Moving Average'], c = 'blue', label = '50-day SMA')
plt.legend()
```

```python
# Step 9a: Create a column containing a 200-day window SMA
new_df['200-day Moving Average'] = new_df['4. close'].rolling(200).mean()
```

```python
# Step 9b: Plot your 50-day and 200-day SMA
plt.figure(figsize = (16,8))
plt.plot(new_df['200-day Moving Average'], c = 'orange', label = '200-day SMA')
plt.plot(new_df['50-day Moving Average'], c = 'blue', label = '50-day SMA')
plt.legend()
```

### Step 10: Drop rows containing empty close_50 and close_200
Since the SMA process leaves the first 200 rows without a price in the column with the 200-day SMA, we will not be able to obtain crosses.


```python
# Step 10: Drop all rows containing NA
new_df.dropna(inplace=True)
new_df
```

### Step 11: Implement a simple SMA crossover strategy

1. You need to keep track of how many share units you have
2. You need to keep track of how much cash you have
3. If close_50 > close_200, you buy all the shares using cash unless you already have shares then you do nothing
4. If close_200 > close_50, you sell all the shares you have and store cash only unless you have only cash then you do nothing
5. At any point in time, you either have only shares or cash, but never both at the same time

```python
# initialize a variable called "shares", and assign 100 to it - this is the number of shares you have at the start
shares = 100

# initialize a variable called "cash", and assign 0 to it - this is how much liquid money you have at the start
cash = 0

# initialize an empty list called "portfolio_sma"
portfolio_sma = []

# initialize an empty list called "cash_list"

cash_list = []
# loop through the DataFrame. Hint: iterrows will be useful here
for index, row in new_df.iterrows():
    # in a particular row, check if "close_50" is more than "close_200"
    if row['50-day Moving Average'] > row['200-day Moving Average']:
        # if "close_50" is more than "close_200", check if "shares" > 0
        if shares > 0:
            # append "shares" * the row's "4. close" price into "portfolio_sma"
            portfolio_sma.append(shares * row['4. close'])
            # append "cash" into "cash_list"
            cash_list.append(cash)
        # else
        else:
            # declare "shares" to be equals to "cash" divided by the row's "4. close"
            shares = cash / row['4. close']
            # declare "cash" to be 0 (because you spent it all buying "shares")
            cash = 0
            # append "shares" * the row's "4. close" price into "portfolio_sma"
            portfolio_sma.append(shares * row['4. close'])
            # append "cash" into "cash_list"
            cash_list.append(cash)
    # else in this case you have "close_200" more than "close_50"
    else:
        # check if you have "shares", i.e. more than 0. Because you have to sell your shares
        if shares > 0:
            # update the value of "cash" to be "shares" multiplied by the row's "4. close"
            cash = shares * row['4. close']
            # update "shares" to be 0 (because you sold all your shares)
            shares = 0
            # append "shares" * the row's "4. close" price into "portfolio_sma"
            portfolio_sma.append(shares * row['4. close'])
            # append "cash" into "cash_list"
            cash_list.append(cash)
        # else
        else:
            # append "shares" * the row's "4. close" price into "portfolio_sma"
            portfolio_sma.append(shares * row['4. close'])
            # append "cash" into "cash_list"
            cash_list.append(cash)

# declare a new column named "portfolio_sma" in your DataFrame which contains "portfolio_sma"
new_df['portfolio_sma'] = portfolio_sma
# declare a new column named "cash_sma" in your DataFrame which contains "cash_list"
new_df['cash_sma'] = cash_list

# check your DataFrame
new_df
```

If you take a look, if you implemented a buy-and-hold strategy, you start off with 65.0 in your portfolio and end up with 66.0 - that's a whopping 1.54% increase. Wow. That's horrible.

On the other hand, if you implement the SMA crossover strategy, you'd end up with 133.52 - <strong>that's a 105.4% increase</strong> instead. You essentially doubled your money in this period.
